# Load data

```python
xs, ys = parse_data('../data/train_x.csv')
train_labels = parse_labels('../data/train_y.csv')
```

```python
data = [(label, x, y) for label, (x, y) in zip(train_labels, zip(xs, ys))]
```

```python
data_label_0 = list(filter(lambda x: x[0] == 0, data))
data_label_1 = list(filter(lambda x: x[0] == 1, data))
data_label_2 = list(filter(lambda x: x[0] == 2, data))
data_label_3 = list(filter(lambda x: x[0] == 3, data))
```

```python
print(len(data_label_0))
print(len(data_label_1))
print(len(data_label_2))
print(len(data_label_3))
```

```python
def create_df(data): 
    x = sum([x for _, x, _ in data], [])
    y = sum([y for _, _, y in data], [])
    return pd.DataFrame(zip(x, y), columns=["x", "y"])

df0 = create_df(data_label_0)
df1 = create_df(data_label_1)
df2 = create_df(data_label_2)
df3 = create_df(data_label_3)
```
